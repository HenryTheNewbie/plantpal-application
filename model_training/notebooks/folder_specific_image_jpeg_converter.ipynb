```python
import os
from PIL import Image

# Function to convert and rename images in each folder
def convert_and_rename_images_in_folder(input_dir, start_index=1):
    # Loop through all folders and files in the input directory
    for root, dirs, files in os.walk(input_dir):
        # Get the folder name for naming convention
        folder_name = os.path.basename(root)

        # Process each file in the folder
        for idx, filename in enumerate(files, start=start_index):
            if filename.endswith((".png", ".jpeg", ".bmp", ".gif", ".tiff", ".jfif", ".jpg", ".JPG")):  # Add other formats if needed
                img_path = os.path.join(root, filename)
                
                # Open the image
                img = Image.open(img_path)

                # Convert the image to RGB mode (necessary for JPEG format)
                rgb_img = img.convert("RGB")

                # Generate new filename using folder name and index starting from 1
                new_filename = f"{folder_name}_{str(idx).zfill(3)}.jpg"
                new_img_path = os.path.join(root, new_filename)

                # Save the image as .jpg with the new filename
                rgb_img.save(new_img_path, "JPEG")
                
                print(f"Converted {filename} to {new_filename}")

                # Delete the original file after conversion
                os.remove(img_path)
                print(f"Deleted original file: {filename}")

# Set your input directory
input_directory = "../data/train_disease/bacterial_spot"  # Adjust this path as necessary

# Run the conversion and renaming, starting from index 1
convert_and_rename_images_in_folder(input_directory, start_index=1)
```
