```python
import os

# Function to display folders and their respective file names
def display_folder_contents(input_dir):
    # Walk through all folders and files in the input directory
    for root, dirs, files in os.walk(input_dir):
        folder_name = os.path.basename(root)  # Get the current folder name
        print(f"Folder: {folder_name}")
        
        # List all files in the folder
        if files:
            for filename in files:
                print(f"  File: {filename}")
        else:
            print("  (No files in this folder)")
        print()  # Blank line for better readability

# Set your input directory
input_directory = "../data/train_disease"  # Adjust this path as necessary

# Run the folder inspection
display_folder_contents(input_directory)
```
